# Interactive Matrix Transformations

This notebook lets you explore how 2x2 matrices transform the plane. You can modify the matrix entries and see how the unit square, unit circle, and individual vectors are transformed.

**No coding required** --- just run the cells and interact with the plots!

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## 1. How a Matrix Transforms the Unit Square

A 2x2 matrix $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$ maps every point $(x, y)$ to a new point $(ax + by, cx + dy)$.

Below, we visualize this by showing what happens to the unit square under different transformations. Try changing the matrix entries!

```python
def plot_transformation(a, b, c, d, title="Matrix Transformation"):
    """Plot the unit square and its image under the matrix [[a, b], [c, d]]."""
    A = np.array([[a, b], [c, d]])
    
    # Unit square vertices
    square = np.array([[0, 0], [1, 0], [1, 1], [0, 1], [0, 0]]).T
    
    # Transform
    transformed = A @ square
    
    # Unit circle
    theta = np.linspace(0, 2 * np.pi, 100)
    circle = np.array([np.cos(theta), np.sin(theta)])
    ellipse = A @ circle
    
    # Basis vectors
    e1 = np.array([[0, 1], [0, 0]])
    e2 = np.array([[0, 0], [0, 1]])
    Ae1 = A @ e1
    Ae2 = A @ e2
    
    fig = make_subplots(rows=1, cols=2, subplot_titles=["Before (Original)", "After (Transformed)"])
    
    # Original
    fig.add_trace(go.Scatter(x=square[0], y=square[1], mode='lines',
                             line=dict(color='blue', width=2), name='Unit square',
                             fill='toself', fillcolor='rgba(68, 114, 196, 0.15)'), row=1, col=1)
    fig.add_trace(go.Scatter(x=circle[0], y=circle[1], mode='lines',
                             line=dict(color='gray', width=1, dash='dot'), name='Unit circle'), row=1, col=1)
    fig.add_trace(go.Scatter(x=e1[0], y=e1[1], mode='lines+markers',
                             line=dict(color='red', width=3), name='e₁',
                             marker=dict(size=8, symbol='arrow', angleref='previous')), row=1, col=1)
    fig.add_trace(go.Scatter(x=e2[0], y=e2[1], mode='lines+markers',
                             line=dict(color='green', width=3), name='e₂',
                             marker=dict(size=8, symbol='arrow', angleref='previous')), row=1, col=1)
    
    # Transformed
    fig.add_trace(go.Scatter(x=transformed[0], y=transformed[1], mode='lines',
                             line=dict(color='blue', width=2), name='Transformed square',
                             fill='toself', fillcolor='rgba(68, 114, 196, 0.15)',
                             showlegend=False), row=1, col=2)
    fig.add_trace(go.Scatter(x=ellipse[0], y=ellipse[1], mode='lines',
                             line=dict(color='gray', width=1, dash='dot'), name='Transformed circle',
                             showlegend=False), row=1, col=2)
    fig.add_trace(go.Scatter(x=Ae1[0], y=Ae1[1], mode='lines+markers',
                             line=dict(color='red', width=3), name='Ae₁',
                             marker=dict(size=8, symbol='arrow', angleref='previous')), row=1, col=2)
    fig.add_trace(go.Scatter(x=Ae2[0], y=Ae2[1], mode='lines+markers',
                             line=dict(color='green', width=3), name='Ae₂',
                             marker=dict(size=8, symbol='arrow', angleref='previous')), row=1, col=2)
    
    det = np.linalg.det(A)
    fig.update_layout(
        title=f"{title}<br><sub>A = [[{a}, {b}], [{c}, {d}]]  |  det(A) = {det:.2f}</sub>",
        height=450, width=900,
        showlegend=True,
        template='plotly_white'
    )
    
    for col in [1, 2]:
        fig.update_xaxes(range=[-3, 3], scaleanchor=f"y{col}", scaleratio=1, row=1, col=col)
        fig.update_yaxes(range=[-3, 3], row=1, col=col)
    
    fig.show()
```

### Identity Matrix
The identity matrix does nothing --- every point maps to itself.

```python
plot_transformation(1, 0, 0, 1, "Identity")
```

### Scaling
A diagonal matrix scales each axis independently.

```python
plot_transformation(2, 0, 0, 0.5, "Scaling: stretch x by 2, compress y by 0.5")
```

### Rotation
A rotation matrix preserves lengths and angles. The determinant is always 1.

```python
angle = np.pi / 6  # 30 degrees
plot_transformation(
    round(np.cos(angle), 3), round(-np.sin(angle), 3),
    round(np.sin(angle), 3), round(np.cos(angle), 3),
    "Rotation by 30\u00b0"
)
```

### Shear
A shear slides one axis along the other. The determinant is 1 (area is preserved).

```python
plot_transformation(1, 1.5, 0, 1, "Horizontal Shear")
```

### Reflection
A reflection flips across an axis. The determinant is $-1$.

```python
plot_transformation(-1, 0, 0, 1, "Reflection across y-axis")
```

### Singular Matrix (Projection)
When the determinant is 0, the matrix collapses space onto a lower-dimensional subspace.

```python
plot_transformation(1, 2, 0.5, 1, "Singular matrix (det = 0): projects onto a line")
```

## 3. Key Takeaways

| Transformation | Matrix form | det(A) | What it preserves |
|---------------|-------------|--------|-------------------|
| Identity | $I$ | 1 | Everything |
| Scaling | $\text{diag}(s_1, s_2)$ | $s_1 \cdot s_2$ | Axis directions |
| Rotation | $R(\theta)$ | 1 | Lengths and angles |
| Shear | Upper/lower triangular with 1s on diagonal | 1 | Area |
| Reflection | Orthogonal with det = $-1$ | $-1$ | Lengths and angles (reverses orientation) |
| Projection | Singular matrix | 0 | Nothing (collapses a dimension) |

The **determinant** tells you how area is scaled: $|\det(A)|$ is the area scaling factor, and the sign tells you whether orientation is preserved (+) or reversed (-).
